```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data = {
    'Year': [2023, 2023, 2023, 2024, 2024, 2024, 2025, 2025],
    'Category': ['Electronics', 'Clothing', 'Health',
                 'Electronics', 'Furniture', 'Groceries',
                 'Clothing', 'Home Goods'],
    'Sales': [250000, 12000, 18000, 30000, 22000, 15000, 22000, 9000],
    'Quantity': [5, 10, 4, 8, 2, 20, 12, 25],
    'Discount': [0.10, 0.20, 0.10, 0.15, 0.05, 0.25, 0.20, 0.30],
    'Profit': [4000, 2000, 1500, 3000, -1000, -500, 3500, -1200]
}

df = pd.DataFrame(data)


yearly_sales = df.groupby('Year')['Sales'].sum()


yearly_sales.plot(kind='bar')
plt.title('Year-wise Total Sales (2023â€“2025)')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.show()

category_sales = df.groupby('Category')['Sales'].sum()


category_sales.plot(kind='bar')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Sales')

plt.show()


category_profit = df.groupby('Category')['Profit'].sum()


category_profit.plot(kind='bar')
plt.axhline(0)
plt.title('Category-wise Profit & Loss')
plt.xlabel('Product Category')
plt.ylabel('Profit / Loss')

plt.show()

correlation = df[['Sales', 'Quantity', 'Discount', 'Profit']].corr()


sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
